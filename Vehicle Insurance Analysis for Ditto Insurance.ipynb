## Feature-Engineering

### Creating New Features

```python
# Create Age Group
df['Age_Group'] = pd.cut(df['Age in Years'], bins=[18, 25, 35, 45, 55, 65, 75, 85], labels=['18-25', '26-35', '36-45', '46-55', '56-65', '66-75', '76-85'])

# Convert Vehicle Age to numerical categories
df['Vehicle_Age_Category'] = df['Vehicle_Age'].map({'< 1 Year': 0, '1-2 Year': 1, '> 2 Years': 2})

# Calculate Premium per Vintage
df['Premium_per_Vintage'] = df['Annual_Premium_cost'] / df['Vintage']

# Create Insurance History
df['Insurance_History'] = df['Previously_Insured_flag'].astype(str) + '_' + df['Vehicle_Damage_boolean']

print("Feature Engineering Done")
```

```python
print(df[['Age_Group', 'Vehicle_Age_Category', 'Premium_per_Vintage', 'Insurance_History']].head())
```

## Encoding

### Encoding Categorical Variables

```python
categorical_features = ['Gender', 'Vehicle_Age', 'Vehicle_Damage_boolean', 'Age_Group', 'Insurance_History']
label_encoder = LabelEncoder()
for feature in categorical_features:
    df[feature] = label_encoder.fit_transform(df[feature])

print("Encoding Done")
```

## Class-Imbalance-Treatment

### Addressing Class Imbalance

```python
X = df.drop(columns=['Response', 'id'])
y = df['Response']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("Class Imbalance Treatment Done")
```

## Standardization

### Standardizing Numerical Features

```python
numeric_features = ['Age in Years', 'Annual_Premium_cost', 'Vintage', 'Premium_per_Vintage']
scaler = StandardScaler()
X_resampled[numeric_features] = scaler.fit_transform(X_resampled[numeric_features])

print("Standardization Done")
```
